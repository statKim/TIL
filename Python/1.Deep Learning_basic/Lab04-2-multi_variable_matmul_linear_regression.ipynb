# Lab 4 Multi-variable linear regression

## Lab04-1-multi_variable_linear_regression
## Using matrix - 앞에서 한 것과 같은 결과

```python
import tensorflow as tf
```

```python
tf.set_random_seed(777)
```

```python
# 5x3 matrix
x_data = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]
# 5x1 vector(matrix)
y_data = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]
```

```python
# placeholder 변수 생성
X = tf.placeholder(tf.float32, shape=[None, 3])     # n x 3 matrix (None : 개수가 정해져있지 않다는 뜻; 데이터의 수)
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
# initial value 지정
W = tf.Variable(tf.random_normal([3, 1]), name='weight')    # weight 3개가 필요!! (3x1)
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
# regression model
hypothesis = tf.matmul(X, W) + b    # tf.matmul(X, W) : 행렬곱(X %*% W) - matrix multiply
```

```python
# cost/loss function (MSE) - matrix form
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
# Session 생성 및 변수 초기화
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train], feed_dict={X: x_data, Y: y_data})
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)
```
